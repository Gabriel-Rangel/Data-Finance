```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pandas_datareader.data as web

import yfinance as yf
yf.pdr_override()
```

```python
# Obtendo as cotações 
ibov = web.get_data_yahoo('^BVSP')
```

```python
ibov.head()
```

```python
ibov.tail()
```

```python
ibov['Close'].plot(figsize=(22,8), label = 'IBOV')
ibov['Close'].rolling(21).mean().plot(label = 'MM21')
ibov['Close'].rolling(200).mean().plot(label = 'MM200')
plt.legend()
```

```python
# Selecionando Período ex.1
ibov_2008 = ibov[ibov.index.year == 2008]
ibov_2008['Close'].plot(figsize=(22,8), label = 'IBOV')
ibov_2008['Close'].rolling(21).mean().plot(label = 'MM21')
ibov_2008['Close'].rolling(200).mean().plot(label = 'MM200')
plt.legend()
```

```python
# Selecionando Período ex. 2
ibov_2008to2016 = ibov[(ibov.index.year >= 2008) & (ibov.index.year <= 2016)]
ibov_2008to2016['Close'].plot(figsize=(22,8), label = 'IBOV')
ibov_2008to2016['Close'].rolling(21).mean().plot(label = 'MM21')
ibov_2008to2016['Close'].rolling(200).mean().plot(label = 'MM200')
plt.legend()
```

```python
# Selecionando Período ex. 3
ibov_last12months = ibov[ibov.index.month == 12]
ibov_last12months ['Close'].plot(figsize=(22,8), label = 'IBOV')
ibov_last12months ['Close'].rolling(21).mean().plot(label = 'MM21')
ibov_last12months ['Close'].rolling(200).mean().plot(label = 'MM200')
plt.legend()

#PS = Tambèm podemos escolher periodo no web.get
#PS1 = O ticker no yahoo fincance tem sufixo .SA
```
